# Fetching movies data from tmbd api and forming a dataset

### Features to be kept from json data ---

- original_title
- popularity
- release_date
- vote_average
- vote_count
- original_language
- id
- overview

### Dataset will be of shape -- 10956 * 8

```python
import pandas as pd
import requests #this library helps to create http requests using python
import os
from dotenv import load_dotenv
```

```python
load_dotenv()

api_key = os.getenv("TMDB_API_KEY")

url = f"https://api.themoviedb.org/3/movie/top_rated?api_key={api_key}&language=en-US&page=1"

response = requests.get(url) #-- response code -- 200 means everything working fine, 404 means error
data = response.json()
print(data)
```
