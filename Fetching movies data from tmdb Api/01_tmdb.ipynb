# Fetching movies data from tmbd api and forming a dataset

### Features to be kept from json data ---

- title
- popularity
- release_date
- vote_average
- vote_count
- original_language
- id
- overview

### Dataset will be of shape -- 10000 * 8 (not 10956 * 8 due to API pagination limit)

```python
import pandas as pd
import requests #this library helps to create http requests using python
import os
from dotenv import load_dotenv
```

```python
load_dotenv()

api_key = os.getenv("TMDB_API_KEY")
```

```python
data_list = [] #initializing an empty list

for i in range(1, 549):

    url = f"https://api.themoviedb.org/3/movie/top_rated?api_key={api_key}&language=en-US&page={i}"
    response = requests.get(url) #response code -- 200(everything well), 404(error)

    if(response.status_code == 200): 

        data = response.json()['results']

    # df = df.append(temp_df, ignore_index= True) in newer version of pandas, append method doesnt exist
    # using df.concat is also slower than the list method

        data_list.extend(data) #dont use data_list.append() as it will create list of lists
    else:
        print(f"API stopped giving data at page {i}. Exiting the loop")
        break

df = pd.DataFrame(data_list)[['id', 'title', 'release_date', 'popularity', 'original_language', 'vote_count', 'vote_average', 'overview']]
df.head(5)

```
