# Fetching movies data from tmbd api and forming a dataset

### Features to be kept from json data ---

- title
- popularity
- release_date
- vote_average
- vote_count
- original_language
- id
- overview

### Dataset will be of shape -- 10956 * 8

```python
import pandas as pd
import requests #this library helps to create http requests using python
import os
from dotenv import load_dotenv
```

```python
load_dotenv()

api_key = os.getenv("TMDB_API_KEY")
```

```python
df = pd.DataFrame() #initializing an empty df

for i in range(1, 549):

    url = f"https://api.themoviedb.org/3/movie/top_rated?api_key={api_key}&language=en-US&page={i}"
    response = requests.get(url) #response code -- 200(everything well), 404(error)

    data = response.json()['results']

    temp_df = pd.DataFrame(data)[['id', 'title', 'release_date', 'popularity', 'original_language', 'vote_count', 'vote_average', 'overview']]

    # df = df.append(temp_df, ignore_index= True) in newer version of pandas, append method doesnt exist
    df = pd.concat([df, temp_df], ignore_index=True)
    
df.head(5)

```
